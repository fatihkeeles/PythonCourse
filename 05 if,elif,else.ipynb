# if , elif , else yapısı

```python
print("Bara 18 yaşından küçükler giremez")

yas = input("Yaşınız kaç : ")

if (yas >= "18"):
    print("Mekana girişiniz serbest...")
```

```python
print("Bara 18 yaşından küçükler giremez")

yas = input("Yaşınız kaç : ")

if (yas >= "18"):
    print("Mekana girişiniz serbest...")
    
else:
    print("18 yaşından küçüksünüz")
```

```python
print("Sınav Geçiş Sistemi")

puan = input("Puanınızı yazınız : ")

if (puan >= "85"):
    print("Sınavdan A aldınız")
elif (puan >= "70"):
    print("Sınavdan B aldınız")

elif (puan >= "50"):
    print("Sınavdan C aldınız")
    
else: 
    print("Sınavdan Geçemediniz")
    
```
